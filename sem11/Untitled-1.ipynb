```python
# x - диапазон -100 100
# y = x*x
from matplotlib import pyplot as plt
import math
def f(x)-> float:
    return math.sin(x)+5*x
x = [x/100 for x in range(-3140, 3140)]
y = [f(t) for t in x]
plt.plot(x, y)
```

```python
from matplotlib import pyplot as plt
import math
def f(x)-> float:
    return 5*x*x+4*x-50
x = [x/100 for x in range(-914, 914)]
y = [f(t) for t in x]
plt.plot(x, y)
```

```python
from sympy import *
from sympy.plotting import plot
```

```python

x = symbols('x')
p1 = plot(x*x -5*x, show = false)
p2 = plot(5*x, show = false)
p1.append(p2[0])
p1.show()
```

```python
x = symbols('x')
p1 = plot(x*x, show = False)
p2 = plot(32*(-x)**0.5, show = False)
p3 = plot(-32*x**0.5, show = False)
p4 = plot(-32*(-x)**0.5, show = False)
p5 = plot(32*x**0.5, show = False)
p6 = plot(-(x*x), show = False)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.append(p5[0])
p1.append(p6[0])
p1.show()
```

```python
from sympy import symbols,sin,cos
from sympy.plotting import plot,plot3d


x = symbols('x')
# plot3d(x-12*x**4*sin(cos(x))-18*x**3+5*x**2 + 10*x - 30,line_color = "purple")
p1 = plot3d(x*x, show = False)
p2 = plot3d(32*(-x)**0.5, show = False)
p3 = plot3d(-32*x**0.5, show = False)
p4 = plot3d(-32*(-x)**0.5, show = False)
p5 = plot3d(32*x**0.5, show = False)
p6 = plot3d(-(x*x), show = False)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.append(p5[0])
p1.append(p6[0])
p1.show()
```
